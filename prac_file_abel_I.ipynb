```python
print("Arithmetic calculator")
print("+-/"*20)
x=int(input("enter first no."))
y=int(input("enter sec. no."))
op=input("enter the operator(*,**,//,/,%,+,-)")
if op=='*':
  res=x*y
elif op=='**':
  res=x**y
elif op=='//':
  res=x//y
elif op=='/':
  res=x/y
elif op=='%':
  res=x%y
elif op=='+':
  res=x+y
elif op=='-':
  res=x-y
else:
  res="invalid operator"
print("^"*30)
print("the result :",x,op,y,"=",res)
```

```python
x=int(input("enter the first no."))
y=int(input("enter sec no."))
z=int(input("enter third no."))
if x<y and x<z:
  if y<z:
    min,mid,max=x,y,z
  else:
    min,mid,max=x,z,y
elif y<x and y<z:
  if x<z:
    min,mid,max=y,x,z
  else:
    min,mid,max=y,z,x
else:
  if x>y:
    min,mid,max=z,y,x
  else:
    min,mid,max=z,y,x
print("the no.s in ascending order is",min,mid,max)
```

```python
num=int(input("enter no. to be reversed"))
rev=0
while num>0:
  rem=num%10
  rev=(rev*10)+rem
  num=num//10
print("reversed no is",rev)
```

```python
num=int(input("enter no."))
rev=0
temp=num
while num>0:
  rem=num%10
  rev=(rev*10)+rem
  num=num//10
if rev==temp:
  print("given no. is palindrome")
else:
  print("given no is not palindrome")
```

```python
print("Armstrong no. checker")
print("-"*30)
num=int(input("enter the no.:"))
print("#"*30)
sum=0
temp=num
while num>0:
  digit=num%10
  sum=sum+digit**3
  num=num//10
if temp==sum:
  print("given no is armstrong")
else:
  print("given no. is not armstrong")
```
